```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
housing
```

```python
df = pd.DataFrame(housing.data)
df
```

```python
df.columns=housing.feature_names
df
```

```python
X=df
y=housing.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=24)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
num_features = X.select_dtypes(exclude="object").columns
numeric_transformer = StandardScaler()
preprocessor = ColumnTransformer(
    transformers=[
        ("numeric",numeric_transformer,num_features)
    ]
)
```

```python
X_train=preprocessor.fit_transform(X_train)
```

```python
X_test=preprocessor.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# model_pipeline = Pipeline(steps=[
#     ('preprocessor', preprocessor),
#     ('model', LinearRegression())
# ])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
models={
    "Linear Regression": LinearRegression()}
```

```python
params ={"Linear Regression": {}}
```

```python
best_models = {}
best_params = {}
best_scores = {}
```

```python
for model_name in models:
    print(f"Performing GridSearchCV for {model_name}...")
    grid_search = GridSearchCV(estimator=models[model_name], param_grid=params[model_name], cv=5, scoring='neg_mean_squared_error', verbose=2)
    grid_search.fit(X_train, y_train)
    best_models[model_name] = grid_search.best_estimator_
    best_params[model_name] = grid_search.best_params_
    best_scores[model_name] = grid_search.best_score_
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
for model_name in best_models:
    y_pred = best_models[model_name].predict(X_test)
    mae, rmse, r2_square = evaluate_model(y_test, y_pred)
    print(f"\n{model_name} Best Parameters: {best_params[model_name]}")
    print(f"{model_name} MAE: {mae}")
    print(f"{model_name} RMSE: {rmse}")
    print(f"{model_name} R² Score: {r2_square}")
```

```python
best_model = LinearRegression()
best_model=best_model.fit(X_train, y_train)
```

```python
best_pred=best_model.predict(X_test)
```
